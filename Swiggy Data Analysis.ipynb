Outliers are data points that are very different (unusually high or unusually low) compared to the rest of the values in the dataset.

# Handling Outliers
In handling outliers we follow 3 methods, They are:

1.Remove Outliers(Deletes rows that fall outside the acceptable range).
Best when:
Dataset is large,Outliers are errors,They add noise to the analysis.

2.Cap outliers
Extreme values are replaced with upper or lower boundaries.
You keep all data points but reduce extreme influence.
- If a value > upper limit → set to upper limit
- If a value < lower limit → set to lower limit

3.Replace outliers with median
Outlier value is replaced with median of the column,Median is stable even when extreme values exist.

and the other method is :
Apply Transformations
Log, sqrt, or Box-Cox transformations reduce skewness.
Makes distribution more normal-like.

# 1.Remove Outliers Method

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df_clean = df[(df[column] >= lower) & (df[column] <= upper)]
    return df_clean
df_no_outliers = remove_outliers_iqr(df, 'Price (INR)')
```

```python
sns.boxplot(x=df['Price (INR)'])
plt.title("Outliers in 'Price (INR)")
plt.show()
```

# observation
This method completely removes rows containing extreme values.
Useful when outliers are due to errors, noise, or extremely rare events.
After removal, boxplots show a tighter distribution, but data loss may occur.

# 2.Cap Outliers Method
Best method compared to others.

```python
col = "Price (INR)"
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)
```

```python
sns.boxplot(x=df['Price (INR)'])
plt.title("Outliers in Price (INR)")
plt.show()
```

# Observation
Outliers are not removed — instead, extreme values are brought to the nearest allowed limit.
This preserves all rows, so no data is lost.
Distribution becomes less skewed because extreme highs/lows are capped.

# 3.Replacing Outliers with Median

```python
def replace_outliers_with_median(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    median_val = df[column].median()
    df[column] = df[column].apply(lambda x: median_val if (x < lower or x > upper) else x)
    return df
df_median = replace_outliers_with_median(df.copy(), 'Price (INR)')
```

```python
df.columns = df.columns.str.strip()
df['Price (INR)'] = pd.to_numeric(df['Price (INR)'], errors='coerce')
df['Rating Count'] = pd.to_numeric(df['Rating Count'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

df_clean = df.dropna(subset=['Price (INR)', 'Rating Count', 'Rating']).copy()

def rank_cut(series, labels):
    n_bins = len(labels)
    ranks = series.rank(method='first')
    total = len(series)
    bin_edges = np.linspace(0, total, n_bins + 1)
    bin_numbers = np.digitize(ranks, bin_edges, right=True)
    bin_numbers[bin_numbers > n_bins] = n_bins  
    return pd.Categorical.from_codes(bin_numbers - 1, labels)

price_labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
rating_labels = ['Very Few', 'Few', 'Moderate', 'Many', 'Very Many']

df_clean['Price_bin'] = rank_cut(df_clean['Price (INR)'], price_labels)
df_clean['RatingCount_bin'] = rank_cut(df_clean['Rating Count'], rating_labels)

heat = df_clean.pivot_table(
    values='Rating',
    index='Price_bin',
    columns='RatingCount_bin',
    aggfunc='mean'
)


plt.figure(figsize=(10,6))
sns.heatmap(heat, annot=True, cmap='Blues', fmt=".2f", cbar_kws={'label': 'Average Rating'})
plt.title("Heatmap: Average Rating by Price & Rating Count Groups", fontsize=14)
plt.xlabel("Rating Count Group")
plt.ylabel("Price Range")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
